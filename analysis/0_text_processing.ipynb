```python
import time

import pandas as pd

import openai
openai.api_key = '' # TODO: private openai api key, removed for sharing purposes
```

```python
def shorten(text, length):
    t = text.split()
    t_len = len(t)
    output = ' '.join(t[:length])
    if length < t_len:
        output += '...'
    return output
```

# review texts

```python
reviews = pd.read_csv('survey_setup_exports/texts/reviews.csv')
```

## title+text joins

```python
# create single-line joined titles+reviews
reviews['text_joined'] = reviews.apply(lambda x: x['r_title']+('' if (x['r_title'][-1]=='.' or x['r_title'][-2]=='.') else '.')+' '+' '.join(x['r_text'].split('\n\n')), axis='columns')
reviews['text_joined'] = reviews['text_joined'].apply(lambda x: ' '.join(x.split()))

reviews[:5]
```

## shorten-100

```python
# create shortened review texts (up to 100 words)
reviews['text_first100'] = reviews.apply(lambda x: shorten(x['text_joined'], 100), axis='columns')
reviews[:5]
```

## export

```python
reviews.to_csv('survey_setup_exports/texts/refined.csv')
```

# bot reviews

## prompt creation

```python
botreviews = pd.DataFrame({'g':[],'m_title':[],'m_yr':[],'prompt':[]})

# create chatgpt-sourced reviews
for g, g_df in reviews.groupby('m_title'):
    example = g_df.iloc[0]
    review_sample_texts = g_df['text_joined']
    prompt = ''
    for i in range(5):
        prompt = prompt+f'\n{(i+1)}. {review_sample_texts.iloc[i]}'
    prompt = (
        'Please write a 75 to 100 word review of this movie, in the style of a friendly recommendation.\n'
        +'Here are some other reviews about this movie to help you out:'
        +prompt
    )
    botreviews = pd.concat([
        botreviews,
        pd.DataFrame({'g':[example['g']],'m_title':[example['m_title']],'m_yr':[example['m_yr']],'prompt':[prompt]})
    ], ignore_index=True)

botreviews[:5]
```

## openai API calls

```python
import pickle
import os

bot_results = {}
bot_result_fname = 'gpt_cached/bot_results.pickle'

if os.path.exists(bot_result_fname):
    with open(bot_result_fname, 'rb') as f:
        bot_results = pickle.load(f)

max_input_length = 4097

for _, r in botreviews.iterrows():
    r_key = (r['m_title'], r['m_yr'])
    if r_key not in bot_results:
        result = openai.ChatCompletion.create(
            model='gpt-3.5-turbo',
            messages=[
                {'role': 'system', 'content': 'You are a friendly movie recommender bot that gives people the most accurate information you can.',},
                {'role': 'user', 'content': r['prompt'][:max_input_length],},
            ],
            # temperature=0,
        )
        time.sleep(20) # to avoid openai rate limit
        bot_results[r_key] = result
        print(f'got result for {r_key}')

with open(bot_result_fname, 'wb') as f:
    pickle.dump(bot_results, f)
```

```python
bot_result_fname = 'gpt_cached/bot_results.pickle'
generated_botreviews = pd.DataFrame({'m_title': [], 'm_yr':[], 'generated':[]})

with open(bot_result_fname, 'rb') as f:
    exporting_bot_results = pickle.load(f)
for k in exporting_bot_results:
    generated_botreviews = pd.concat([
        generated_botreviews,
        pd.DataFrame({
            'm_title':[k[0]],
            'm_yr':[k[1]],
            'generated':[exporting_bot_results[k]['choices'][0]['message']['content']]
        })
    ], ignore_index=True)
#     print(shorten(v['choices'][0]['message']['content'], 100))
#     print('---===---===---===---===---===---===---')
generated_botreviews[:5]
```

```python
# join generated text with prompts

merged_botreviews = botreviews.merge(generated_botreviews, how='inner', on=['m_title', 'm_yr'])
merged_botreviews[:5]
```

```python
# create shortened review texts (up to 100 words)
merged_botreviews['generated_first100'] = merged_botreviews.apply(lambda x: shorten(x['generated'], 100), axis='columns')
merged_botreviews[:5]
```
