# Validate workerIDs

```python
def extract_args(url):
    url_args = url[url.index('?')+1:].split('&')
    return dict([tuple(e.split('=')) for e in url_args])

# for each of the mturk entries, verify that they entered their workerID into Qualtrics and that they entered the responseID into MTurk
validated_ids = pd.DataFrame()
for _, row in df_mturk.iterrows():
    row_args = extract_args(row['Input.SURVEY_LINK'])
    matched = df_qualt.loc[
        (df_qualt['workerid']==row['WorkerId'])
        & (df_qualt['ResponseId']==row['Answer.surveycode'])
        & (
            (df_qualt['param_x']==row_args['param_x'])
            & (df_qualt['param_a']==row_args['param_a'])
            & (df_qualt['param_b']==row_args['param_b'])
            & (df_qualt['param_c']==row_args['param_c'])
            & (df_qualt['param_d']==row_args['param_d'])
            & (df_qualt['param_e']==row_args['param_e'])
        )
        & (df_qualt['Progress']=='100')
    ]
    if matched.shape[0]==1:
        validated_ids = pd.concat([validated_ids, pd.DataFrame({
            'WorkerID': [row['WorkerId']],
            'HitID': [row['HITId']],
            'AssignmentID': [row['AssignmentId']],
            'ResponseID': [row['Answer.surveycode']],
            'Progress': [int(matched.iloc[0]['Progress'])],
            'time_sec': [int(matched.iloc[0]['Duration (in seconds)'])],
            'paid_bonus': ['-'],
        })])
    else:
        print('invalid info combination from MTurk:', row['WorkerId'], row['HITId'], row['Answer.surveycode'])

validated_ids = validated_ids.reset_index()
validated_ids.to_csv('VALIDATED_IDS.csv', sep='\t')
validated_ids
```

```python
import boto3
import math

# Add qualification / tag to workers who have been approved for this task / already did it once before

### TRUE iff actually marking qualification payment. FALSE iff estimating
live_qual = False

# hyperparameter : the ID of the 'completed this survey before' qualification
qual_id = 'TODO'

# using [https://boto3.amazonaws.com/v1/documentation/api/latest/reference/services/mturk/client/list_workers_with_qualification_type.html]
# using [https://boto3.amazonaws.com/v1/documentation/api/latest/reference/services/mturk/client/associate_qualification_with_worker.html]
mturk = boto3.client(
    'mturk',
    endpoint_url=endpoint_url,
    region_name=region_name,
)

# first quietly BUILD the list of workers who are already qualified, so we can mark if there were any duplicate responders
already_qualified = []
response = mturk.list_workers_with_qualification_type(
    QualificationTypeId=qual_id,
    Status='Granted',
    MaxResults=100,
)
while response['NumResults'] > 0:
    already_qualified = already_qualified + [
        e['WorkerId'] for e in response['Qualifications']
    ]
    response = mturk.list_workers_with_qualification_type(
        QualificationTypeId=qual_id,
        Status='Granted',
        NextToken=response['NextToken'],
        MaxResults=100,
    )
print(f'already have {len(already_qualified)} qualified workers from before')
print()

print(f'MARKING QUALS FOR {len(validated_ids)} WORKERS...')
print(f'(simulating quals = {not live_qual}) // (actual quals = {live_qual})')
num_duplicate = 0
for _, row in validated_ids.iterrows():
    worker_id = row['WorkerID']
    print()
    print(worker_id, qual_id)
    duplicate_qual = False
    if worker_id in already_qualified:
        print('worker is already qualified?')
        num_duplicate += 1
        duplicate_qual=True
    if live_qual and not duplicate_qual:
        response = mturk.associate_qualification_with_worker(
            QualificationTypeId=qual_id,
            WorkerId=worker_id,
            IntegerValue=1,
            SendNotification=False,
        )
        print(response)
print()
print(f'total {num_duplicate} duplicated responders')
```

# Send MTurk bonuses

```python
import boto3
import math

### TRUE iff actually sending payment. FALSE iff calculating payment
live_send = False

# using [https://boto3.amazonaws.com/v1/documentation/api/latest/reference/services/mturk/client/send_bonus.html]
mturk = boto3.client(
    'mturk',
    endpoint_url=endpoint_url,
    region_name=region_name,
)

total_bonus_pay = 0
print(f'SENDING BONUSES FOR {len(validated_ids)} CASES...')
print(f'(simulating sends = {not live_send}) // (actual sends = {live_send})')
for _, row in validated_ids.iterrows():
    hit_id = row['HitID']
    # calculate 15/hour payment rate
    time_sec = row['time_sec']
    time_min = time_sec/60
    time_hour = time_min/60
    totalpay = time_hour*15
    bonuspay = totalpay-2 # deduct the approval payment that is already paid
    bonuspay = math.ceil(bonuspay*100)/100
    # make sure we do not have any negative bonus amounts
    bonuspay = max([bonuspay, 0])
    # cap out at maximum bonus amount
    bonuspay = min([bonuspay, 5.50])
    if bonuspay>0:
        # build bonus request
        bonus_request = {
            'WorkerId': row['WorkerID'],
            'AssignmentId':  row['AssignmentID'],
            'BonusAmount': str(bonuspay),
            'Reason': f'Thank you for participating in our study (HIT {hit_id})!',
            'UniqueRequestToken': (row['WorkerID']+'_'+str(bonuspay)+'_'+row['AssignmentID'])[:63],
        }
        print()
        print(row['WorkerID'], row['time_sec'], bonuspay)
        print(bonus_request)
        total_bonus_pay += bonuspay
        if live_send:
            response = mturk.send_bonus(**bonus_request)
            print(response)
    else:
        print()
        print('skipping pay for this case:', row['WorkerID'], row['time_sec'])
print()
print(f'total estimated bonus pay not including fees = {total_bonus_pay}')
```

```python
# using [https://boto3.amazonaws.com/v1/documentation/api/latest/reference/services/mturk/client/list_bonus_payments.html#MTurk.Client.list_bonus_payments]
mturk = boto3.client(
    'mturk',
    endpoint_url=endpoint_url,
    region_name=region_name,
)

print('SHOWING SENT BONUSES...')
for _, row in validated_ids.iterrows():
    hit_id = row['HitID']
    # calculate 15/hour payment rate
    time_sec = row['time_sec']
    time_min = time_sec/60
    time_hour = time_min/60
    totalpay = time_hour*15
    bonuspay = totalpay-0.5 # deduct the approval payment that is already paid
    bonuspay = math.ceil(bonuspay*100)/100
    unique_request_token = (row['WorkerID']+'_'+str(bonuspay)+'_'+row['AssignmentID'])[:63]
    past_bonus_request = {
        'HITId': hit_id,
        # 'AssignmentId':  row['AssignmentID'],
    }
    print()
    print(row['WorkerID'], bonuspay, unique_request_token)
    print(past_bonus_request)
    response = mturk.list_bonus_payments(**past_bonus_request)
    print(response)
```
